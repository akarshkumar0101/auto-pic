```python
class ImageCPPN(nn.Module):
    def __init__(self, n_hidden=24, n_layers=8, n_channels=3, activation=torch.relu, norm=False, res=False, features=['y', 'x', 'r'], cache_coordinates=True):
        super().__init__()
        
        self.norm=norm
        self.features = features
        self.res = res
        self.idx_features_to_use = []
        feature2idx = {'y': 0, 'x': 1, 'r': 2, 'theta': 3, 'ay': 4, 'ax': 5}
        for feature in self.features:
            print(feature, feature2idx[feature])
            self.idx_features_to_use.append(feature2idx[feature])

        self.cache_coordinates = {} if cache_coordinates else None
        
        n_units = [len(self.idx_features_to_use)] + [n_hidden]*n_layers + [n_channels]
        n_ins = np.array(n_units[:-1])
        n_outs = np.array(n_units[1:])
        
        # print(n_ins, n_outs)
        self.layers = nn.ModuleList([nn.Conv2d(n_in, n_out, kernel_size=1) 
                                     for n_in, n_out in zip(n_ins, n_outs)])
        # print(self.layers)
        
        self.activation = activation
        
    def forward(self, x):
        x = x[:, self.idx_features_to_use, :, :]

        for layer in self.layers[:-1]:
            lx = layer(x)
            if self.norm:
                lx = lx/lx.mean(dim=(-1,-2), keepdim=True)
                # lx = lx/lx.std(dim=(-1,-2), keepdim=True)
                # lx = (lx - lx.mean(dim=(-1,-2), keepdim=True))/lx.std(dim=(-1, -2), keepdim=True)
                # lx = lx/lx.mean(dim=(-3), keepdim=True)
            lx = self.activation(lx)
            
            x = x+lx if self.res and x.shape==lx.shape else lx
            
        x = self.layers[-1](x)
        x = x.sigmoid()
        return x

    def generate_input(self, img_size=(224, 224)):
        device = self.layers[0].weight.device

        if self.cache_coordinates is not None and img_size in self.cache_coordinates:
            return self.cache_coordinates[img_size].to(device)

        h, w = img_size
        y, x = torch.linspace(-1, 1, h, device=device), torch.linspace(-1, 1, w, device=device)
        y, x = torch.meshgrid(y, x, indexing='ij')
        
        r = (x.pow(2)+y.pow(2)).sqrt()
        theta = torch.atan2(x, y)
        # costheta = torch.cos(theta)
        # plt.imshow(to_np(theta));plt.show()
        # plt.imshow(to_np(costheta));plt.show()
        ay, ax = y.abs(), x.abs()
        
        data = [y, x, r, theta, ay, ax]
        x = torch.stack(data, dim=0)[None] # 1, 6, h, w
        # normalize
        x = (x-x.mean(dim=(-1, -2), keepdim=True))/x.std(dim=(-1, -2), keepdim=True)

        if self.cache_coordinates is not None:
            self.cache_coordinates[img_size] = x

        return x

    def generate_image(self, img_size=(224, 224)):
        x = self.generate_input(img_size)
        return self(x)
```

```python
torch.manual_seed(4)
cppn = ImageCPPN(activation=torch.tanh, norm=True, res=True)
# cppn = ImageCPPN(activation=torch.relu, norm=True, res=True)
cppn = cppn.to(device)

target = torch.from_numpy(malik).permute(2, 0, 1).to(device)

opt = torch.optim.Adam(cppn.parameters(), lr=1e-3)

for i in tqdm(range(10000)):
    img = cppn.generate_image(img_size=target.shape[-2:])
    loss = (img-target).pow(2).mean()
    
    opt.zero_grad()
    loss.backward()
    opt.step()

    if i%300==0:
        print("Loss: ", loss.item())
        plt.imshow(to_np(img[0].permute(1, 2, 0)))
        
        plt.show()
```
